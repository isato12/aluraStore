### Importación de datos



```python
import pandas as pd
from matplotlib import pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda.Producto
```

#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

# Sumar los ingresos totales de todas las tiendas
ingreso_total_general = ingreso_total_tienda1 + ingreso_total_tienda2 + ingreso_total_tienda3 + ingreso_total_tienda4

print(f"Ingreso total Tienda 1: ${ingreso_total_tienda1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_total_tienda2:,.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_total_tienda3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_total_tienda4:,.2f}")
print(f"\nIngreso total general de todas las tiendas: ${ingreso_total_general:,.2f}")
```

```python
# Crear una lista con los nombres de las tiendas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear una lista con los ingresos totales de cada tienda
ingresos_totales = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]

# Crear el gráfico de barras
plt.bar(nombres_tiendas, ingresos_totales)
plt.ylabel('Ingreso Total ($)')
plt.title('Ingreso Total por Tienda')
plt.xticks(rotation=0)  # Mantener las etiquetas horizontales para mayor claridad
# Añadir las cantidades a las barras
for i, v in enumerate(ingresos_totales):
    plt.text(i, v, f"${v:,.2f}", ha='center', va='top', rotation=90, color='white', fontsize=9, weight='bold')
plt.show()
```

# 2. Ventas por categoría

```python
venta_por_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].mean().round(2)
venta_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].mean().round(2)
venta_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].mean().round(2)
venta_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].mean().round(2)

# Sumar las Series para obtener el promedio total por categoría
ingreso_total_general_por_categoria = (venta_por_categoria_tienda1 + venta_por_categoria_tienda2 + venta_por_categoria_tienda3 + venta_por_categoria_tienda4).round(2)

print("Ingreso total de la Tienda 1 por categoria es:")
display(venta_por_categoria_tienda1)
print("\nIngreso total de la Tienda 2 por categoria es:")
display(venta_por_categoria_tienda2)
print("\nIngreso total de la Tienda 3 por categoría es:")
display(venta_por_categoria_tienda3)
print("\nIngreso total de la Tienda 4 por categoría es:")
display(venta_por_categoria_tienda4)
print("\nIngreso total general (suma de promedios) de todas las tiendas por categoría:")
display(ingreso_total_general_por_categoria)
```

```python
# Crear un DataFrame combinado con los promedios por categoría de cada tienda
ventas_por_categoria_combinado = pd.DataFrame({
    'Tienda 1': venta_por_categoria_tienda1,
    'Tienda 2': venta_por_categoria_tienda2,
    'Tienda 3': venta_por_categoria_tienda3,
    'Tienda 4': venta_por_categoria_tienda4
})

# Crear el gráfico de barras agrupadas
ax = ventas_por_categoria_combinado.plot(kind='bar', figsize=(12, 6))
plt.ylabel('Precio Promedio ($)')
plt.title('Precio Promedio por Categoría y Tienda')
plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mayor claridad
plt.tight_layout() # Ajustar el layout para evitar que las etiquetas se solapen

# Añadir las cantidades a las barras
for container in ax.containers:
    ax.bar_label(container, fmt='{:,.2f}', rotation=90, padding=5)


plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

# Concatenar los DataFrames para calcular la calificación promedio general
calificacion_total = pd.concat([tienda['Calificación'], tienda2['Calificación'], tienda3['Calificación'], tienda4['Calificación']])
calificacion_promedio_general = calificacion_total.mean()


print(f"calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
print(f"\ncalificación promedio general de todas las tiendas: {calificacion_promedio_general:.2f}")
```

```python
# Crear una lista con los nombres de las tiendas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear una lista con los ingresos totales de cada tienda
calificacion_promedio = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]

# Crear el gráfico de barras

plt.bar(nombres_tiendas, calificacion_promedio)
plt.ylabel('calificación promedio')
plt.title('Calificación promedio por tienda')
plt.xticks(rotation=0)  # Mantener las etiquetas horizontales para mayor claridad

# Añadir las cantidades a las barras
for i, v in enumerate(calificacion_promedio):
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center')

plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_vendidos_tienda1 = tienda['Producto'].value_counts()
productos_vendidos_tienda2 = tienda2['Producto'].value_counts()
productos_vendidos_tienda3 = tienda3['Producto'].value_counts()
productos_vendidos_tienda4 = tienda4['Producto'].value_counts()

# Concatenar las Series y sumar los conteos para obtener el total general por producto
productos_vendidos_total = pd.concat([productos_vendidos_tienda1, productos_vendidos_tienda2, productos_vendidos_tienda3, productos_vendidos_tienda4]).groupby(level=0).sum()


print("Conteo de productos vendidos en la Tienda 1:")
display(productos_vendidos_tienda1.head(3)) # Mostrar los primeros 3 para no llenar la salida
print("\nConteo de productos vendidos en la Tienda 2:")
display(productos_vendidos_tienda2.head(3))
print("\nConteo de productos vendidos en la Tienda 3:")
display(productos_vendidos_tienda3.head(3))
print("\nConteo de productos vendidos en la Tienda 4:")
display(productos_vendidos_tienda4.head(3))
print("\nConteo total de productos vendidos en todas las tiendas:")
display(productos_vendidos_total.head(3))

print("\nProducto más vendido en la Tienda 1:")
display(productos_vendidos_tienda1.idxmax())
print("\nProducto menos vendido en la Tienda 1:")
display(productos_vendidos_tienda1.idxmin())

print("\nProducto más vendido en la Tienda 2:")
display(productos_vendidos_tienda2.idxmax())
print("\nProducto menos vendido en la Tienda 2:")
display(productos_vendidos_tienda2.idxmin())

print("\nProducto más vendido en la Tienda 3:")
display(productos_vendidos_tienda3.idxmax())
print("\nProducto menos vendido en la Tienda 3:")
display(productos_vendidos_tienda3.idxmin())

print("\nProducto más vendido en la Tienda 4:")
display(productos_vendidos_tienda4.idxmax())
print("\nProducto menos vendido en la Tienda 4:")
display(productos_vendidos_tienda4.idxmin())

print("\nProducto más vendido en total:")
display(productos_vendidos_total.idxmax())
print("\nProducto menos vendido en total:")
display(productos_vendidos_total.idxmin())
```

```python
# Crear un DataFrame combinado con los promedios por categoría de cada tienda
productos_mas_vendidos = pd.DataFrame({
    'Tienda 1': productos_vendidos_tienda1.head(3),
    'Tienda 2': productos_vendidos_tienda2.head(3),
    'Tienda 3': productos_vendidos_tienda3.head(3),
    'Tienda 4': productos_vendidos_tienda4.head(3)
})

# Crear el gráfico de barras agrupadas
productos_mas_vendidos.plot(kind='bar', figsize=(12, 6))
plt.ylabel('Productos')
plt.title('Productos más vendidos')
plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mayor claridad
plt.tight_layout() # Ajustar el layout para evitar que las etiquetas se solapen
plt.show()

# Productos menos vendidos por tienda

productos_menos_vendidos = pd.DataFrame({
    'Tienda 1': productos_vendidos_tienda1.tail(3),
    'Tienda 2': productos_vendidos_tienda2.tail(3),
    'Tienda 3': productos_vendidos_tienda3.tail(3),
    'Tienda 4': productos_vendidos_tienda4.tail(3)
})

productos_menos_vendidos.plot(kind='bar', figsize=(12, 6))
plt.ylabel('Cantidad Vendida')
plt.title('Productos Menos Vendidos por Tienda')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Producto más vendido en total
plt.figure(figsize=(8, 5))
productos_vendidos_total.head(1).plot(kind='bar')
plt.ylabel('Cantidad Vendida')
plt.title('Producto Más Vendido en General')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
